```python
import pandas as pd

# Загрузка данных из CSV-файла
df = pd.read_csv('/home/qqshnik/Documents/labs/AI/Beer Profile and Ratings Data Set/beer_profile_and_ratings.csv')

print(df.head())
```

```python
print("Data type: ", type(df))
print("Data shape: ", df.shape)
print(df.dtypes)
```

```python
# Функция для обрезки значения до символа '-' (не включая его)
def trim_style(style):
    if '-' in style:
        return style.split('-')[0].strip()  # Обрезаем до '-' и убираем пробелы
    return style  # Если '-' нет, возвращаем исходное значение

# Применяем функцию к столбцу 'Style'
df['Style'] = df['Style'].apply(trim_style)

# Получение уникальных значений из столбца 'Style'
unique_styles = df['Style'].unique()

# Вывод уникальных значений
print("Сорта пива (Уникальные значения в столбце без разделения на регионы и тд.)")
for style in unique_styles:
    print(style)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))

for style in unique_styles:
    style_data = df[df['Style'] == style]
    plt.scatter(style_data['Min IBU'], style_data['Max IBU'], label=style)

# Настройка графика
plt.xlabel('Min IBU')
plt.ylabel('Max IBU')
plt.title('Min IBU vs Max IBU by Beer Style')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.grid(True)
plt.tight_layout()

# Показать график
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Считаем количество каждого стиля
style_counts = df['Style'].value_counts()

# Строим столбчатую диаграмму
style_counts.plot(kind='bar', color='skyblue')

# Настройка графика
plt.xlabel('Style')
plt.ylabel('Count')
plt.title('Распределение стилей пива')
plt.xticks(rotation=45, ha='right')  # Поворачиваем подписи для удобства чтения
plt.tight_layout()

plt.figure(figsize=(14, 8))

# Строим boxplot для ABV по стилям
df.boxplot(column='ABV', by='Style', figsize=(14, 8), grid=False, rot=90)

# Настройка графика
plt.xlabel('Beer Style')
plt.ylabel('ABV (Alcohol by Volume)')
plt.title('Alcohol Content by Beer Style')
plt.suptitle('')  # Убираем автоматический заголовок "Boxplot grouped by Style"
plt.tight_layout()

# Показать третий график
plt.show()
```
