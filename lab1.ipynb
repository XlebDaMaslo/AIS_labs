```python
import pandas as pd

# Загрузка данных из CSV-файла
df = pd.read_csv('./data/beer_profile_and_ratings.csv')

print(df.head())
```

```python
print("Data type: ", type(df))
print("Data shape: ", df.shape)
print(df.dtypes)
```

```python
# Функция для обрезки значения до символа '-' (не включая его)
def trim_style(style):
    if '-' in style:
        return style.split('-')[0].strip()  # Обрезаем до '-' и убираем пробелы
    return style  # Если '-' нет, возвращаем исходное значение

# Применяем функцию к столбцу 'Style'
df['Style'] = df['Style'].apply(trim_style)

# Получение уникальных значений из столбца 'Style'
unique_styles = df['Style'].unique()

# Вывод уникальных значений
print("Сорта пива (Уникальные значения в столбце без разделения на регионы и тд.)")
for style in unique_styles:
    print(style)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))

for style in unique_styles:
    style_data = df[df['Style'] == style]
    plt.scatter(style_data['Min IBU'], style_data['Max IBU'], label=style)

# Настройка графика
plt.xlabel('Min IBU')
plt.ylabel('Max IBU')
plt.title('Min IBU vs Max IBU by Beer Style')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.grid(True)
plt.tight_layout()

# Показать график
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))

# Считаем количество каждого стиля
style_counts = df['Style'].value_counts()

style_counts.plot(kind='bar', color='skyblue')

plt.xlabel('Style')
plt.ylabel('Count')
plt.title('Распределение стилей пива')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Содержание алкоголя по стилям пива (Violin Plot)
plt.figure(figsize=(14, 8))

sns.violinplot(x='Style', y='ABV', hue='Style', data=df, inner='quartile', palette='Set3')

plt.xlabel('Style')
plt.ylabel('ABV (Alcohol by Volume)')
plt.title('Содержание алкоголя по стилям пива')
plt.xticks(rotation=90)
plt.ylim([0, 20]) # Единичные высокие значения мешают визуальному анализу  
plt.tight_layout()

# Горечь пива по стилям (Min IBU и Max IBU)
# Группируем данные по стилям и вычисляем средние значения Min IBU и Max IBU
style_ibu = df.groupby('Style')[['Min IBU', 'Max IBU']].mean().reset_index()

# Сортируем данные по среднему значению Min IBU для удобства
style_ibu = style_ibu.sort_values(by='Min IBU', ascending=False)

plt.figure(figsize=(14, 8))

# Ширина столбцов
bar_width = 0.35

# Позиции для столбцов
positions = range(len(style_ibu))

# Строим столбцы для Min IBU
plt.bar(positions, style_ibu['Min IBU'], width=bar_width, color='skyblue', label='Min IBU')

# Строим столбцы для Max IBU
plt.bar([p + bar_width for p in positions], style_ibu['Max IBU'], width=bar_width, color='lightcoral', label='Max IBU')

# Настройка графика
plt.xlabel('Beer Style')
plt.ylabel('IBU (International Bitterness Units)')
plt.title('Горечь пива по стилям (Min IBU и Max IBU)')
plt.xticks([p + bar_width / 2 for p in positions], style_ibu['Style'], rotation=90)  # Подписи по центру столбцов
plt.legend()
plt.tight_layout()

# Показать график
plt.show()
```
