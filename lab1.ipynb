```python
import pandas as pd

df = pd.read_csv('./data/beer_profile_and_ratings.csv')

print("Тип данных: ", type(df))
print("Форма данных: ", df.shape)
print("Тип данных каждого признака:")
print(df.dtypes)
```

```python
print("\nКоличество пропущенных значений в каждом столбце:")
print(df.isnull().sum())

original_rows = df.shape[0]
df = df.dropna(axis=0, how='any')

print(f"Удалено строк: {original_rows - df.shape[0]}")
print(f"Новый размер данных: {df.shape}")
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import chi2

print("\nАнализ выбросов с использованием расстояния Махаланобиса:")

# Выбираем числовые признаки для анализа
numeric_features = ['ABV', 'Min IBU', 'Max IBU']

# Визуализация распределения до обработки
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_features], orient="h", palette="Set2")
plt.title("Распределение признаков до обработки")
plt.show()

# Функция для вычисления расстояния Махаланобиса
def mahalanobis_dist(df, features):
    X = df[features].values
    mean = np.mean(X, axis=0)
    cov = np.cov(X, rowvar=False)
    
    try:
        cov_inv = np.linalg.inv(cov)
    except np.linalg.LinAlgError:
        cov_inv = np.linalg.pinv(cov)  # Используем псевдообратную матрицу если обычная вырождена
    
    deviations = X - mean
    distances = np.sqrt(np.sum(deviations.dot(cov_inv) * deviations, axis=1))
    return distances

# Вычисляем расстояния
distances = mahalanobis_dist(df, numeric_features)

# Определяем порог с использованием хи-квадрат распределения
threshold = chi2.ppf(0.95, df=len(numeric_features))  # 95% доверительный интервал

# Находим индексы выбросов
outliers_mask = distances > np.sqrt(threshold)
outliers = df[outliers_mask]

print(f"\nОбнаружено многомерных выбросов: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")

# Визуализация расстояний
plt.figure(figsize=(10, 6))
sns.histplot(distances, bins=50, kde=True)
plt.axvline(x=np.sqrt(threshold), color='r', linestyle='--', label='Порог')
plt.title("Распределение расстояний Махаланобиса")
plt.xlabel("Расстояние")
plt.legend()
plt.show()

# Удаляем выбросы
original_size = len(df)
df_clean = df[~outliers_mask]

print(f"\nУдалено строк: {original_size - len(df_clean)} ({(original_size - len(df_clean))/original_size*100:.2f}%)")
print(f"Новый размер датасета: {df_clean.shape}")

# Визуализация после удаления
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean[numeric_features], orient="h", palette="Set2")
plt.title("Распределение после удаления выбросов")
plt.show()

# Обновляем основной датафрейм
df = df_clean.copy()
```

```python
# Функция для обрезки значения до символа '-' (не включая его)
def trim_style(style):
    if '-' in style:
        return style.split('-')[0].strip()
    return style

df['Style'] = df['Style'].apply(trim_style)

unique_styles = df['Style'].unique()

# Вывод уникальных значений
print("Сорта пива (Уникальные значения в столбце без градаций внутри стилей)")
for style in unique_styles:
    print(style)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

for style in unique_styles:
    style_data = df[df['Style'] == style]
    plt.scatter(style_data['Min IBU'], style_data['Max IBU'], label=style)

# Настройка графика
plt.xlabel('Min IBU')
plt.ylabel('Max IBU')
plt.title('Min IBU vs Max IBU by Beer Style')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.grid(True)
plt.tight_layout()

# Показать график
plt.show()
```

```python
# Обновленная классификация по ABV
def classify_abv(abv):
    """Классификация пива по содержанию алкоголя (ABV)"""
    if abv <= 0.5:
        return "Безалкогольное (ABV ≤ 0,5%)"
    elif 0.5 < abv <= 4:
        return "Слабоалкогольное (0,5 - 4% ABV)"
    elif 4 < abv <= 5.5:
        return "Стандартное (4–5,5% ABV)"
    elif 5.5 < abv <= 10:
        return "Крепкое (5,5–10% ABV)"
    elif 10 < abv <= 20:
        return "Очень крепкое (10–20% ABV)"
    else:
        return "Экстремальные сорта (ABV >20%)"

# Применяем классификацию к данным
df['ABV_Class'] = df['ABV'].apply(classify_abv)

# Визуализация с классификацией по ABV
plt.figure(figsize=(10, 10))

# Используем seaborn для scatter plot с цветовой классификацией
sns.scatterplot(
    x='Min IBU', 
    y='Max IBU', 
    hue='ABV_Class',  # Цвет точек по классификации ABV
    data=df, 
    palette='tab10',  # Яркая и контрастная палитра
    s=100,  # Размер точек
    alpha=0.8  # Прозрачность
)

# Настройка графика
plt.xlabel('Min IBU')
plt.ylabel('Max IBU')
plt.title('Горечь пива (Min IBU vs Max IBU) с классификацией по ABV')
plt.legend(title='Классификация ABV', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Показать график
plt.show()
```

```python
from sklearn.cluster import KMeans

# Признаки для кластеризации (ABV и Avg IBU)
df['Avg IBU'] = (df['Min IBU'] + df['Max IBU']) / 2
X = df[['ABV', 'Avg IBU']]

n_clusters = 7  # Количество кластеров
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(15, 10))

sns.scatterplot(x='ABV', y='Avg IBU', hue='Cluster', data=df, palette='viridis', s=100)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Cluster Centers')

plt.xlabel('ABV (Alcohol by Volume)')
plt.ylabel('Average IBU (International Bitterness Units)')
plt.title('Кластеризация пива по содержанию алкоголя и горечи')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()

plt.show()
```
